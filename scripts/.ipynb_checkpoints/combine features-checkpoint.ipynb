```python
import pandas as pd
```

```python
%cd ..
```

```python
face = pd.read_csv('output/features/face_detection_results.csv', index_col = 0)
face.head()
```

```python
color = pd.read_csv('output/features/color_results.csv', index_col = 0)
color.head()
```

```python
car = pd.read_csv('output/features/cars_results.csv', index_col = 0)
car = car.reset_index()
car.head()
```

```python
# remove duplicates
color = color.drop_duplicates('path', keep = False)[['id', 'median_h','median_s','median_v','std_h','std_s','std_v', 'binary']]
face = face.drop_duplicates('path', keep = False)[['id', 'face']]
car = car[['id', 'has_cars']]
```

```python
merged = face.merge(color, how='inner', on='id')
merged = merged.merge(car, how='inner', on='id')

merged.face = merged.face.astype('int64')
merged.binary = merged.binary.astype('int64')
merged.head()
```

```python
merged.to_csv('output/full_features.csv')
```
